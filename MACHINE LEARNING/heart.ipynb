```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your dataset
df = pd.read_csv("heart.csv")   # change filename if needed

# Separate features (X) and target (y)
X = df.drop("HeartDisease", axis=1)   # replace 'target' if your column name is different
y = df["HeartDisease"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% test data
    random_state=42,    # same result every time
    stratify=y          # keeps class balance (important for classification)
)

# Check result
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```
