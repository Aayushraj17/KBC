```python
one_hot_enododing = pd.get_dummies(x,columns=[ 'iso3','region','biome','rodent_host_species','exposure_setting' ],drop_first=True)
```

```python
one_hot_enododing
```

```python
one_hot_enododing = one_hot_enododing.astype(int)
```

## now doing label encoding 

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

columns = ['iso3','region','biome','rodent_host_species','exposure_setting' ]
xlabel = x.copy()
label_encoder = {}

for col in columns:
    le = LabelEncoder()
    xlabel[col] = le.fit_transform(xlabel[col].astype(str)) #convert to str into null values 
    
    label_encoder[col] = le
    
```

```python
xlabel
```

## DOING STANDARD SCALING TO CONVERT LARGE NUMERIC NUMBER INTO SMALLE ONE WITH THE HELP OF ONE_HOT_ENCODING

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_only =['latitude','longitude','year','quarter','avg_temp_c','rainfall_mm','humidity_pct','ndvi','deforestation_km2']

scaler = StandardScaler()

one_hot_enododing[numeric_only] = scaler.fit_transform(one_hot_enododing[numeric_only])
```

```python
one_hot_enododing.head()
```

```python
xlabel[['iso3', 'region', 'latitude', 'longitude', 'biome', 'year', 'quarter',
       'avg_temp_c', 'rainfall_mm', 'humidity_pct', 'ndvi',
        'rodent_host_species', 'deforestation_km2',
       'exposure_setting']] = scaler.fit_transform(xlabel[['iso3', 'region', 'latitude', 'longitude', 'biome', 'year', 'quarter',
       'avg_temp_c', 'rainfall_mm', 'humidity_pct', 'ndvi',
        'rodent_host_species', 'deforestation_km2',
       'exposure_setting']])
```
