## now doing label encoding to convert str into numeric one 

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

columns = ['brand', 'model', 'fuel_type', 'transmission', 
       'owner_type', 'city', ]
xlabel = x.copy()
label_encoder = {}

for col in columns:
    le = LabelEncoder()
    xlabel[col] = le.fit_transform(xlabel[col].astype(str)) #convert to str into null values 
    
    label_encoder[col] = le
    
```
