## doing standart  scaling convert large numeric into smaller one with the help of one hot encoding

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_colum = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement',
       'yr_built', 'yr_renovated']

scaler = StandardScaler()

one_hot_encoding[numeric_colum] = scaler.fit_transform(one_hot_encoding[numeric_colum])


```

```python
one_hot_encoding
```

## creating model with the help of one_hot_encoding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(one_hot_encoding, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

## doing predection

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_pred, y_test)
```
