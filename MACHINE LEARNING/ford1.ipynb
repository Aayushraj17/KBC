```python
df = df.drop_duplicates(subset=['model', 'year', 'price', 'transmission', 'mileage', 'fuelType', 'tax',
       'mpg', 'engineSize'])
```

```python
x = df.drop(['price'], axis = 1)
y = df['price']
```

## APPLYING ONE HOT ENCODING TO CONVERRT STR INTO NUMERIC TERM 


```python
one_hot_encoding = pd.get_dummies(x,columns=['model','transmission','fuelType'],drop_first=True)
```

```python
one_hot_encoding = one_hot_encoding.astype(int)
```

```python
one_hot_encoding
```

```python
one_hot_encoding.info()
```

## NOW DOING LABEL ENCODING 

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

columns = ['model','transmission','fuelType']
xlabel = x.copy()
label_encoder = {}

for col in columns:
    le = LabelEncoder()
    xlabel[col] = le.fit_transform(xlabel[col].astype(str)) #convert to str into null values 
    
    label_encoder[col] = le
    
```

```python
xlabel
```

## DOING STANDARD SCLAING WITH THE HELP OF ONE_HOT_ENCODING

```python
one_hot_encoding.columns

```

```python
from sklearn.preprocessing import StandardScaler

```

```python
numeric_col = ['year','mileage','tax','mpg','engineSize']
scaler = StandardScaler()

one_hot_encoding[numeric_col] = scaler.fit_transform(one_hot_encoding[numeric_col])
```

```python
one_hot_encoding
```

## NOW DOING  XLABEL-ENCODING TO CONVERT LARGE NUMBER INTO SMALL NUMERIC 

```python
xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
xlabel
```

## CREATING MODEL WITH THE HELP OF ONE_HOT_ENCODING

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(one_hot_encoding, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_pred,y_test)
```
